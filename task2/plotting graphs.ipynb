```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from numpy import genfromtxt
data_11= genfromtxt('csvs/run_logs-cifar10_task11-tag-accuracy.csv', delimiter=',')
data_12 = genfromtxt('csvs/run_logs-cifar10_task12-tag-accuracy.csv', delimiter=',')[0::2,:]
data_13 = genfromtxt('csvs/run_logs-cifar10_task13-tag-accuracy.csv', delimiter=',')[0::2,:]
data_14 = genfromtxt('csvs/run_logs-cifar10_task14-tag-accuracy.csv', delimiter=',')
```

```python
plt.plot(data_11[1:,1], np.array([data_11[1:,2], data_12[1:,2], data_13[1:,2], data_14[1:,2]]).T)
plt.legend(['no augmentation' ,'normalized', 'flipping', 'cropping'])
plt.xlabel('iterations')
plt.ylabel('training accuracy')
plt.savefig('graphs/task1.eps')
```
